```python
from dotenv import load_dotenv
import os
from pyairtable import Api
import pandas as pd
```

```python
load_dotenv()
KEY = os.getenv('KEY')
APP = os.getenv('APP')
TBL = os.getenv('TBL')
api = Api(KEY)
```

```python

table = api.table(APP, TBL).all()
```

```python
# df = [field.get("fields") for field in table]
df = pd.DataFrame(table)
df
```
