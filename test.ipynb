```python
from dotenv import load_dotenv
import os
from pyairtable import Api
from models import *
import pandas as pd


load_dotenv()
KEY = os.getenv('KEY')
APP = os.getenv('APP')

api = Api(KEY)
base = api.base(APP)
```

```python
subject_table = api.table(APP, "Subjects")
subject_table
```
